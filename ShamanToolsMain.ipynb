```
Start Program

Display "Bienvenidos a la tienda Shaman Tools"
Display "1. Iniciar sesión"
Display "2. Salir"

Read user_option

If user_option is "2"
    End Program

If user_option is "1"
    Call login function
    If login successful
        current_user = {username, role}
        If current_user["role"] is "admin"
            Call manage_admin function
        Else if current_user["role"] is "seller"
            Call manage_seller function
    Else
        Display "Error de autenticación, opción no válida."
        Return to welcome message

Else
    Display "Opción inválida"
    Return to welcome message


// Función de autenticación de usuario
Function login
    Display "Ingrese nombre de usuario:"
    Read username
    Display "Ingrese contraseña:"
    Read password
    If username exists in users and password matches
        current_user = {username, role}
        Return True
    Else
        Return False


// Menu del administrador
Function manage_admin
    While True
        Display "Menú Admin: 1. Registrar Usuario, 2. Añadir/Actualizar Producto, 3. Ver Inventario, 4. Buscar Producto, 5. Comprar Producto, 6. Resumir Ventas, 7. Cerrar sesión"
        Read admin_option
        If admin_option is "1"
            Call create_user function
        If admin_option is "2"
            Call add_update_product function
        If admin_option is "3"
            Call list_products function
        If admin_option is "4"
            Call search_product function
        If admin_option is "5"
            Call buy_product function
        If admin_option is "6"
            Call list_sales function
        If admin_option is "7"
            Call logout function
            Exit manage_admin


// Menú del vendedor
Function manage_seller
    While True
        Display "Menú Vendedor: 1. Ver Inventario, 2. Buscar Producto, 3. Vender Producto, 4. Cerrar sesión"
        Read seller_option
        If seller_option is "1"
            Call list_products function
        If seller_option is "2"
            Call search_product function
        If seller_option is "3"
            Call sell_product function
        If seller_option is "4"
            Call logout function
            Exit manage_seller


// Funciones de operaciones
Function create_user
    Display "Ingrese nombre de usuario, contraseña y rol"
    Register new user in users

Function add_update_product
    Display "Ingrese nombre_producto, precio_compra, precio_venta, stock y tipo_unidad"
    Add or update product in products

Function list_products
    Display productos en products con detalles

Function search_product
    Display "Ingrese código del producto"
    If product found in products
        Display detalles del producto
    Else
        Display "Producto no encontrado"

Function buy_product
    Display "Ingrese código del producto y cantidad a comprar"
    Increase product stock in products

Function list_sales
    Display ventas registradas en sales con detalles

Function sell_product
    Display "Ingrese código del producto y cantidad a vender"
    If sufficient stock available
        Decrease stock in products
        Add sale to sales

Function logout
    current_user = {}
    Return to main menu
```
